```python
import pandas as pd
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/s4sauravv/Datasets/main/Admission_Prediction.csv")
df.head()
```

```python
# printing  the shape of the dataframe

df.shape
```

```python
# counting the total values in the dataset

df.count()
```

```python
# finding the null values in the dataset

df.isnull().sum()
```

```python
# filling the null values
df['GRE Score']=df['GRE Score'].fillna(df['GRE Score'].mean())
df['TOEFL Score']=df['TOEFL Score'].fillna(df['TOEFL Score'].mean())
df['University Rating']=df['University Rating'].fillna(df['University Rating'].mode()[0])
```

```python
df.isnull().sum()
```

Now we have zero null values

```python
# describing the data and finding some details

df.describe()
```

Lets delet the columns which are not use to us

```python
df.drop(columns=['Serial No.'], inplace=True)
df.head()
```

#### Now lets do some data visualization and analyse the with the help of charts

```python
%matplotlib inline
import matplotlib.pyplot as plt
df.hist(bins=50, figsize=(20, 15))
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(10,6),facecolor='white')
plotnumber=1
for column in df:
    if plotnumber <=8:
        axs=plt.subplot(2,4,plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column,fontsize=12)
    plotnumber +=1
plt.tight_layout()
```

```python
# splitting the data into features and outputs

y=df['Chance of Admit']
x=df.drop(columns=['Chance of Admit'])
```

```python
plt.figure(figsize=(10,6),facecolor='white')
plotnumber=1
for column in x:
    if plotnumber <=8:
        axs=plt.subplot(2,4,plotnumber)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=12)
        plt.ylabel('Chance of Admit',fontsize=12)
        plt.title('Featurs VS Label')
    plotnumber +=1
plt.tight_layout()
```

#### Now we convert Standard Scaler 

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_scaler=scaler.fit_transform(x)
x_scaler
```

```python
#Now do Train test split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test =train_test_split(x_scaler,y,test_size=0.25,random_state=(94255))
```

```python
# Creating an instances of the model
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
df.tail(2)
```

```python
lr.predict(scaler.transform([[327.0,113.0,4.0,4.5,4.5,9.04,0]]))
```

```python
from sklearn.metrics import r2_score

lr.score(x_train, y_train)
```

```python
lr.score(x_test,y_test)
```

```python
x_graph=lr.predict(x_test)
```

```python
plt.scatter(y_test,x_graph)
plt.xlabel('Actual Data')
plt.ylabel('Predicted Data')
plt.title('Linear Regression Graph')
plt.tight_layout()
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
#MAE
mean_absolute_error(y_test,x_graph)
```

```python
#MSE
mean_squared_error(y_test,x_graph)
```

```python
import numpy as np

np.sqrt(mean_squared_error(y_test,x_graph))
```
